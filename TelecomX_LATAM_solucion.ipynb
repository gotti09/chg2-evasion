```python
# Informe de Análisis de Evasión de Clientes (Churn) – TelecomX

## 1. Introducción

La evasión de clientes, conocida como **Churn**, es uno de los principales desafíos para las empresas de telecomunicaciones, ya que implica la cancelación del servicio por parte de los clientes. Este fenómeno tiene un impacto directo en los ingresos y en la sostenibilidad del negocio, dado que adquirir nuevos clientes suele ser más costoso que retener a los existentes.

El objetivo de este análisis es **explorar y comprender los factores asociados a la evasión de clientes en la empresa TelecomX**, utilizando técnicas de análisis de datos. A partir de un conjunto de datos que contiene información demográfica, características del servicio contratado, cargos y comportamiento de los clientes, se busca identificar patrones que expliquen por qué ciertos clientes tienen mayor probabilidad de cancelar el servicio.

El análisis permitirá generar **insights que ayuden a la empresa a tomar decisiones estratégicas para mejorar la retención de clientes y reducir la tasa de churn**.

---

## 2. Limpieza y Tratamiento de Datos

El proceso de preparación de los datos se realizó siguiendo un flujo de trabajo típico de **ETL (Extracción, Transformación y Carga)**.

### 2.1 Extracción de datos

Los datos fueron obtenidos desde una API pública en formato JSON mediante Python. Para ello se utilizaron las librerías `requests` y `pandas`, lo que permitió descargar y convertir los datos a una estructura tabular.

Posteriormente, se utilizó la función `json_normalize()` para **aplanar las estructuras anidadas del JSON**, convirtiendo los diferentes grupos de información (cliente, cuenta, internet, teléfono) en columnas dentro de un DataFrame.

### 2.2 Transformación y limpieza

Durante esta etapa se realizaron diversas tareas de preparación de datos:

* Conversión de datos a tipos adecuados para su análisis.
* Verificación de valores únicos en variables categóricas.
* Identificación de valores faltantes o inconsistentes.
* Creación de una nueva variable denominada **“Cuentas_Diarias”**, calculada a partir del cargo mensual del cliente dividido entre 30 días.
* Exploración de distribuciones mediante frecuencias y proporciones para detectar posibles anomalías en los datos.

Estas transformaciones permitieron **preparar un conjunto de datos limpio y adecuado para el análisis exploratorio**.

### 2.3 Carga de datos

Finalmente, el conjunto de datos procesado fue almacenado en un archivo Excel para facilitar su consulta y reutilización en análisis posteriores.

---

## 3. Análisis Exploratorio de Datos (EDA)

El análisis exploratorio tuvo como objetivo **comprender la estructura de los datos e identificar patrones asociados a la evasión de clientes**.

### 3.1 Estadísticas descriptivas

Se calcularon métricas estadísticas utilizando la función `describe()` de pandas, incluyendo:

* Media
* Mediana
* Desviación estándar
* Valores mínimos y máximos
* Cuartiles

Estas métricas permitieron comprender la distribución de variables numéricas como:

* Tiempo de permanencia del cliente
* Cargos mensuales
* Cargos totales
* Cuentas diarias

### 3.2 Análisis de variables categóricas

Se analizaron diferentes variables categóricas para observar su relación con la evasión de clientes. Entre ellas:

* Género del cliente
* Tipo de contrato
* Método de pago
* Tipo de servicio de internet
* Presencia de pareja o dependientes

Para cada variable se calcularon **proporciones de churn** mediante tablas cruzadas (`crosstab`) con normalización, lo que permitió identificar diferencias entre grupos de clientes.

### 3.3 Identificación de patrones

El análisis permitió detectar patrones relevantes, tales como:

* Los clientes con **contratos mensuales** presentan mayores tasas de cancelación.
* Algunos **métodos de pago electrónicos** están asociados a mayor evasión.
* Ciertos **servicios de internet** presentan tasas de churn más elevadas.
* Los clientes con **menor tiempo de permanencia** tienden a cancelar con mayor frecuencia.

Estos patrones ayudan a identificar segmentos de clientes con mayor riesgo de abandono.

---

## 4. Conclusiones e Insights

A partir del análisis realizado, se pueden destacar los siguientes hallazgos principales:

1. **El tipo de contrato es uno de los factores más influyentes en la evasión de clientes.**
   Los clientes con contratos mes a mes presentan tasas significativamente mayores de cancelación en comparación con aquellos que tienen contratos anuales o de largo plazo.

2. **El método de pago también influye en la probabilidad de churn.**
   Algunos métodos de pago muestran mayor proporción de cancelaciones, lo que podría estar asociado a perfiles de clientes más inestables.

3. **El tiempo de permanencia del cliente está relacionado con la lealtad.**
   Los clientes nuevos presentan mayor riesgo de cancelar el servicio.

4. **Las características del servicio contratado pueden afectar la satisfacción del cliente.**
   Diferencias en los tipos de servicios de internet o paquetes contratados pueden influir en la decisión de cancelar.

En conjunto, estos resultados muestran que **la evasión de clientes no ocurre de forma aleatoria**, sino que está relacionada con características específicas del cliente y del servicio.

---

## 5. Recomendaciones

A partir de los resultados del análisis, se proponen las siguientes recomendaciones estratégicas:

### 1. Incentivar contratos de mayor duración

Ofrecer descuentos o beneficios adicionales a los clientes que opten por contratos de uno o dos años podría ayudar a **reducir significativamente la tasa de cancelación**.

### 2. Fortalecer estrategias de retención temprana

Implementar programas de fidelización para clientes nuevos durante sus primeros meses de servicio puede disminuir la probabilidad de abandono.

### 3. Analizar la experiencia del cliente según el servicio contratado

Evaluar la calidad del servicio de internet y otros servicios asociados puede ayudar a detectar posibles problemas que generen insatisfacción.

### 4. Implementar modelos predictivos de churn

Con los datos disponibles, la empresa podría desarrollar **modelos de predicción de evasión** para identificar clientes con alto riesgo de cancelar el servicio y aplicar acciones preventivas.

### 5. Personalizar ofertas y comunicaciones

Segmentar a los clientes según sus características y comportamientos permitirá ofrecer promociones o soluciones más adecuadas a cada perfil.

---

## Conclusión general

El análisis de datos realizado permitió identificar **factores clave asociados a la evasión de clientes en TelecomX**. Mediante el uso de técnicas de exploración y análisis estadístico fue posible detectar patrones relevantes que pueden orientar estrategias de retención.

La aplicación de estas recomendaciones, junto con un monitoreo continuo del comportamiento de los clientes, puede contribuir a **reducir la tasa de churn y mejorar la rentabilidad del negocio**.
```
